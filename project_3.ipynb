```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dengue= pd.read_csv("dengue_datasets\\dengue .csv")
```

```python
dengue
```

```python
dengue.groupby("Year")["Dengue_Cases"].sum().plot(kind="bar")
plt.title("Total Dengue Cases per Year")
plt.xlabel("Year")
plt.ylabel("Total Cases")
plt.show()
```

<h1>1. Total Dengue Cases per Year</h1>
<h3>The bars show which years had the highest dengue cases. If one year (like 2019) is taller, it means dengue outbreaks were stronger that year compared to others</h3>

```python
dengue.groupby("Year")["Dengue_Deaths"].sum().plot(kind="bar")
plt.title("Total Dengue Deaths per Year")
plt.xlabel("Year")
plt.ylabel("Total Deaths")
plt.show()
```

<h1>2. Total Dengue Deaths per Year</h1>
<h3>A year with a high death count suggests a severe outbreak or weaker response in healthcare. Years with fewer deaths show better dengue control.</h3>

```python
dengue.groupby("Region")["Dengue_Cases"].mean().plot(kind="bar")
plt.title("Average Dengue Cases per Region")
plt.xlabel("Region")
plt.ylabel("Average Cases")
plt.show()
```

<h1>3. Average Dengue Cases per Region</h1>
<h2>Regions with taller bars experienced more dengue cases on average. This indicates hotspots where dengue prevention is most needed.</h2>

```python
dengue.groupby("Region")["Dengue_Deaths"].sum().plot(kind="bar")
plt.title("Total Dengue Deaths per Region")
plt.xlabel("Region")
plt.ylabel("Total Deaths")
plt.show()
```

<h1>4. Total Dengue Deaths per Region</h1>
<h2>Regions with higher total deaths likely faced more severe outbreaks or had less access to medical help.</h2>

```python
dengue.groupby("Month")["Dengue_Cases"].sum().plot(kind="line", marker="o")
plt.title("Monthly Dengue Cases Trend (All Years)")
plt.xlabel("Month")
plt.ylabel("Cases")
plt.show()
```

<h1>5. Monthly Dengue Trend (All Years)</h1>
<h2>Peaks in certain months (like June to October) suggest dengue cases rise during the rainy season, when mosquitoes breed more easily.</h2>

```python
plt.scatter(dengue["Dengue_Cases"], dengue["Dengue_Deaths"])
plt.title("Dengue Cases vs Deaths")
plt.xlabel("Dengue Cases")
plt.ylabel("Dengue Deaths")
plt.show()
```

<h1>6. Dengue Cases vs Deaths</h1>
<h2>The scatter shows that as dengue cases increase, deaths also rise â€” proving a positive relationship between the two.</h2>

```python
dengue.boxplot(column="Dengue_Cases", by="Year")
plt.title("Distribution of Dengue Cases by Year")
plt.suptitle("")
plt.xlabel("Year")
plt.ylabel("Cases")
plt.show()
```

<h1>7. Distribution of Dengue Cases by Year</h1>
<h2>The boxplot reveals which years had a wider spread of cases. A large range means dengue levels fluctuated more during that year.</h2>

```python
dengue.boxplot(column="Dengue_Deaths", by="Year")
plt.title("Distribution of Dengue Deaths by Year")
plt.suptitle("")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.show()
```

<h1>8. Distribution of Dengue Deaths by Year</h1>
<h2>Similar to cases, this shows years with inconsistent or extreme death counts â€” possibly due to outbreaks or health response differences.</h2>

```python
pivot_cases = dengue.pivot_table(values="Dengue_Cases", index="Year", columns="Region", aggfunc="sum")
pivot_cases.plot()
plt.title("Dengue Cases per Region Over Time")
plt.xlabel("Year")
plt.ylabel("Cases")
plt.show()
```

<h1>9. Dengue Cases per Region Over Time</h1>
<h2>The lines show how dengue cases rise and fall by region across years. Some regions may show steady increases, others declines</h2>

```python
pivot_deaths = dengue.pivot_table(values="Dengue_Deaths", index="Year", columns="Region", aggfunc="sum")
pivot_deaths.plot()
plt.title("Dengue Deaths per Region Over Time")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.show()
```

<h1>10. Dengue Deaths per Region Over Time</h1>
<h2>Trends show which regions consistently had high death numbers, pointing to areas needing stronger intervention</h2>

```python
dengue.groupby("Year")[["Dengue_Cases", "Dengue_Deaths"]].sum().plot(kind="bar")
plt.title("Comparison of Dengue Cases and Deaths per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

<h1>11.Comparison of Dengue Cases and Deaths per Year</h1>
<h2>The bars show both numbers together even though deaths are lower, they still follow the same trend as total cases.</h2>

```python
dengue.groupby("Month")["Dengue_Cases"].sum().sort_values(ascending=False).head(5).plot(kind="bar")
plt.title("Top 5 Months with Highest Dengue Cases")
plt.xlabel("Month")
plt.ylabel("Cases")
plt.show()
```

<h1>12. Top 5 Months with Highest Dengue Cases</h1>
<h2>These top months reveal dengueâ€™s â€œdanger season.â€ Focused awareness and cleanup drives during these months can help reduce spread.</h2>

```python
dengue.groupby(["Year", "Month"])["Dengue_Cases"].mean().unstack(0).plot()
plt.title("Monthly Average Dengue Cases per Year")
plt.xlabel("Month")
plt.ylabel("Average Cases")
plt.show()
```

<h1>13. Monthly Average Dengue Cases per Year</h1>
<h2>chart compares how monthly dengue patterns changed from year to year.</h2>

```python
dengue.groupby("Month")["Dengue_Deaths"].mean().plot(kind="bar")
plt.title("Average Dengue Deaths per Month")
plt.xlabel("Month")
plt.ylabel("Average Deaths")
plt.show()
```

<h1>14. Average Dengue Deaths per Month</h1>
<h2>Shows which months are deadlier, often matching the same months where cases are high, emphasizing seasonal risk.</h2>

```python
(dengue.groupby("Region")["Dengue_Deaths"].sum() / dengue.groupby("Region")["Dengue_Cases"].sum()).plot(kind="bar")
plt.title("Death-to-Case Ratio per Region")
plt.xlabel("Region")
plt.ylabel("Ratio")
plt.show()
```

<h1>15. Death-to-Case Ratio per Region</h1>
<h2>Regions with higher ratios have more deaths compared to cases â€” possibly due to healthcare access issues or delayed treatment.</h2>
